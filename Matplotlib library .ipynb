```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8,5,9])
x2=np.array([5,8,7,16])
plt.plot(x,marker="o",ms=20,mec="green",ls='dashed',lw=7,c='hotpink')
plt.plot(x2,marker="o",ms=20,mec="yellow",ls='dotted',lw=5,c='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8,5,9])
x2=np.array([5,8,7,16])
plt.plot(x,marker="o",ms=20,mec="green",ls='dashed',lw=7,c='hotpink')
plt.plot(x2,marker="o",ms=20,mec="yellow",ls='dotted',lw=5,c='red')
plt.xlabel("age")
plt.ylabel("weight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8,5,9])
x2=np.array([5,8,7,16])
plt.plot(x,marker="o",ms=20,mec="green",ls='dashed',lw=7,c='hotpink')
plt.plot(x2,marker="o",ms=20,mec="yellow",ls='dotted',lw=5,c='red')
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age_weight graph",loc='right')
plt.grid(axis='x',ls="--",lw=7,c='green')
plt.show()
```
