Map Function

Python's map() is a built-in function that allows you to process and transform all the items in an iterable without using an explicit for loop, a technique commonly known as mapping.

```python
def even_odd(num):
    if num%2==0:
        return True
    else:
        return False
```

```python
even_odd(25)
```

```python
def even_odd(num):
    if num%2==0:
        return "the number {}  is even ".format(num)
    else:
        return "the number {} is odd  ".format(num)
```

```python
even_odd(58)
```

```python
lst=[1,2,4,3,79,5,7,8,9,988,7,65,4]
```

```python
map(even_odd,lst)
```

```python
list(map(even_odd,lst))
```

```python
t=(1,3,2,4,2,3,4,66,7,88,9,5)
```

```python
tuple(map(even_odd,t))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Adjusted range to be inclusive
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print("The number is prime")
else:
    print("The number is not prime")
```

```python
lst=[1,2,3,4,5,6,7,77,6,55,43,23]
```

```python
list(map(is_prime,lst))
```

Recursion Function:

The process in which a function calls itself directly or indirectly is called recursion and the corresponding function is called a recursive function.

Recursion means you are calling a function from itself.


```python
def greet():
    print("hello")
    greet() #this function calls itself again and again
greet()
```

```python
import sys
sys.setrecursionlimit(100)
print(sys.getrecursionlimit())
try:
    def greet():
        print("hello")
        greet() #this function itself again and again
except:
    greet()
```

```python

def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
rec=fact(5)
print(rec)
```

```python
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
num=int(input("enter the no you want to calculate recursion= "))
fact(num)
```

Advantages of using recursion

A complicated function can be split down into smaller sub-problems utilizing recursion.

Sequence creation is simpler through recursion than utilizing any nested iteration.

Recursive functions render the code look simple and effective.

Disadvantages of using recursion

A lot of memory and time is taken through recursive calls which makes it expensive for use.

Recursive functions are challenging to debug.

              |
              
    func() ----

Fibanacci Series

```python
def feb(n):
    if n<=1:
        return n
    else:
        return (feb(n-1)+feb(n-2))
term=10
if term <=0:
    print("invalid syntax!")
else:
    print("the feb series is given below")
for i in range(term):
    print(feb(i))
```

```python
def fib(n):
    if n<=1:
        return n
    else:
        return(fib(n-1)+fib(n-2))
num=int(input("enter the number of terms you want to calculate fibonacci series = "))
if num<=0:
    print("enter a valid number ")
else:
    print("the possible terms for the given no is given below :")
for i in range(num):
    print(fib(i))
```
